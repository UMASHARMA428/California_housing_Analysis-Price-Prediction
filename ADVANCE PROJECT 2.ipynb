# Ques 5:
Create a data set by filling the missing data with the mean value of the total_bedrooms in the original data set.

```python
import pandas as pd

# Create new dataset by filling missing values with the mean
df_copy = df.copy()  # copy the data of df to new dataframe name df_copy
df_copy["total_bedrooms"]= df_copy["total_bedrooms"].fillna(df_copy['total_bedrooms'].mean())

# to display the  new data set
print(df_copy)

# In above code,a new dataset(df_copy) created where the missing values in 'total_bedrooms' denoted by NaN
# are replaced with mean value of 'total_bedrooms'.
```

# Ques 10:
Please create a new column named total_bedroom_size. If size 10 or less. It should be quoted as small. If the size is 11 or more but 1000 less, it should be medium, otherwise it should be large.

```python
import numpy as np
#conditions
a = [(df['total_bedrooms'] <=10),
      (df['total_bedrooms'] >=11)&(df['total_bedrooms'] <=1000),
          (df['total_bedrooms'] > 1000)]

label = ['small','medium','large']

df['total_bedroom_size'] =np.select(a,label) #np.select fx return an array drawn from elements in choicelist, depending on conditions.

print(df.head())

# Total bedroom size :- Ordinal type as it have categories with a meaningful order, but the intervals btwn categories r not uniform.

# 'total_bedrooms' :- Discrete type, as it represents a count of bedrooms and takes on distinct, separate values.
```
